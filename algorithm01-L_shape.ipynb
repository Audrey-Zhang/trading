```python
# distribute one k to scale
def distri_k_to_scale(kDf, scaleL, countL):
    count_times = 0.0
    for scale in scaleL:
        if kDf.L <= scale and scale <= kDf.H:
            count_times += 1
    count_unit = tm_unit_value / count_times

    for j,scale in enumerate(scaleL):
        print j,scale,countL[j],':',
        if kDf.L <= scale and scale <= kDf.H:
            countL[j] += count_unit
            print 'count:' + str(countL[j])
        else:
            print 'count:' + str(countL[j])
    return True
```

```python
# update distribution on a stick at every new k
# columns = ['id','Tm','L','H','scale','count','sd']
def update_scale_distri(kDf,cur_distriL):
    L = min(cur_distriL['L'] ,kDf.L)
    H = max(cur_distriL['H'] ,kDf.H)
    if L != cur_distriL['L'] or H != cur_distri['H']:  # if HL changed: update scale
        internal = ( H - L ) / cur_distriL['sd']
        cur_distriL['scale'] = np.arange(L,H,internal)
        print 'scale list change:',kDf.Tm
        
    distri_k_to_scale(kDf,cur_distriL['scale'],cur_distriL['count'])
    
    return True
```
