```python
class StdK:
    def __init__(self, **kwargs):
        self.drt = kwargs['drt']
        if 'k_bar' in kwargs.keys():
            self.H, self.L, self.TmIdx = kwargs['k_bar'][1], kwargs['k_bar'][2], kwargs['k_bar'][4]
        if 'H' in kwargs.keys():
            self.H, self.L, self.TmIdx = kwargs['H'], kwargs['L'], kwargs['TmIdx']
    
    def update(self, k_bar):
        print('a:', k_bar)
        flag = 0 # 0:stain; 1: up new; -1: down new
        if (k_bar[1] > self.H and k_bar[2] > self.L) or (self.drt == 1 and k_bar[1] > self.H):
            flag = 1
            k_bar[2] = max(k_bar[2], self.L)
        elif (k_bar[1] < self.H and k_bar[2] < self.L) or (self.drt == -1 and k_bar[2] < self.L):
            flag = -1
            k_bar[1] = min(k_bar[1], self.H)
        else:
            flag = 0
            self.H = min(self.H, k_bar[1])
            self.L = max(self.L, k_bar[2])
        print('b:', k_bar)
        return flag, k_bar 
    
    def update_T(self, H, L, TmIdx):
        flag = 0 # 0:stain; 1: up new; -1: down new
        new_stdK = {'H':H, 'L':L, 'drt':flag, 'TmIdx': TmIdx}
        if (H > self.H and L > self.L) or (self.drt == 1 and H > self.H):
            flag = 1
            new_stdK['L'] = max(L, self.L)
        elif (H < self.H and L < self.L) or (self.drt == -1 and L < self.L):
            flag = -1
            new_stdK['H'] = min(H, self.H)
        else:
            flag = 0
            self.H = min(self.H, H)
            self.L = max(self.L, L)
        return flag, new_stdK 

                
```

```python
class Stick:
    def __init__(self, method, **kwargs):
        if method == 'init':
            self.TmIdx_start, self.TmIdx_end = kwargs['k_bar'][4], kwargs['k_bar'][4]
            self.O, self.C = kwargs['k_bar'][0], kwargs['k_bar'][3]
            self.start = Point(kwargs['k_bar'][4], kwargs['k_bar'][0], drt=0)
            self.std_k_bar_stack = [StdK(k_bar=kwargs['k_bar'], drt=0)]
            self.peak = self.getPoint(0, 'init')
            self.peak_position = 0
            self.drt = 0
            
            self.reverse_count = 0
            self.status = 0 # 0:open; 1:ripe; 2:close
        elif method == 'trim':
            self.std_k_bar_stack = kwargs['std_k_bar_stack']
            self.TmIdx_start, self.TmIdx_end = self.std_k_bar_stack[0].TmIdx, self.std_k_bar_stack[-1].TmIdx
            self.drt = self.std_k_bar_stack[1].drt
            self.peak_position = kwargs['peak_position']
            if self.drt == 1:
                self.O, self.C = self.std_k_bar_stack[0].L, self.std_k_bar_stack[-1].H
                self.start = getPoint(0, 'L')
                self.peak = getPoint(self.peak_position, 'H')
            elif self.drt == -1:
                self.O, self.C = self.std_k_bar_stack[0].H, self.std_k_bar_stack[-1].L
                self.start = getPoint(0, 'H')
                self.peak = getPoint(self.peak_position, 'L')
            self.reverse_count = 1
            self.status = 1 # 0:open; 1:ripe; 2:close
    
    def update(self, k_bar):
        self.TmIdx_end, self.C = k_bar[4], k_bar[3]
        flag = 0 # 0:stain; 1: updated; 2:cloes
        new_stick = ''
        if self.updateStdK(k_bar) != 0: # stdK, stdK_stackstdKStick
            peak, pp = self.findPeak()
            if peak == 0:
                flag = self.updateRange(k_bar)
            elif self.status == 0:
                self.status == 1
                self.peak_position == pp
                self.drt = -peak
                self.reverse_count = 1
            elif peak != self.drt: # 
                if self.drt == 1 and self.std_k_bar_stack[-2].H >= self.peak_value:
                    self.peak_position = pp
                    self.reverse_count = 1
                    flag = 1                    
                elif self.drt == -1 and self.std_k_bar_stack[-2].L <= self.peak_value:
                    self.peak_position = pp
                    self.reverse_count = 1
                    flag = 1 
            elif peak == self.drt and self.reverse_count > 3:
                self.closeStick()
                new_stick = self.new_stick(pp)
                flag = 2                        
        return flag, new_stick
    
    def updateStdK(self, k_bar, drt=0):
        is_new = 0
        is_new, trimed_k_bar = self.std_k_bar_stack[-1].update(k_bar)
        if is_new != 0:
            self.std_k_bar_stack.append(StdK(k_bar=trimed_k_bar, drt=drt)) 
            self.reverse_count += 1
        return is_new
            
    def save(self):
        return None
    
    #=================================================================================
    def findPeak(self):
        peak_type = 0
        if self.reverse_count > 2:
            if (self.std_k_bar_stack[-2].H > self.std_k_bar_stack[-1].H and 
               self.std_k_bar_stack[-2].H > self.std_k_bar_stack[-3].H):
                peak_type = -1
            elif (self.std_k_bar_stack[-2].L < self.std_k_bar_stack[-1].L and 
               self.std_k_bar_stack[-2].L < self.std_k_bar_stack[-3].L):
                peak_type = 1
        return peak_type, len(self.std_k_bar_stack) - 2
    
    def updateRange(self, k_bar):
        flag = 0  # 0: stick stain; 1: stick updated
        if self.drt == 1 and k_bar[1] >= self.peak.V:
            self.peak = Point(k_bar[4], k_bar[1], -1)
            flag = 1
        elif self.drt == -1 and k_bar[2] <= self.peak.V:
            self.peak = Point(k_bar[4], k_bar[2], 1)
            flag = 1
        return flag
    
    def closeStick(self):
        self.status = 2
        return None
    
    def new_Stick(self, peak_position_in_std_stack):
        std_k_bar_stack = self.std_k_bar_stack[self.peak_position_in_std_stack:]
        peak_position = len(std_k_bar_stack) - 2

        new_stick = {
            'std_k_bar_stack' : std_k_bar_stack,
            'peak_position' : peak_position
        }
        return new_stick
    
    def getPoint(self, position_in_std_stack, method):
        if method == 'H':
            p = Point(self.std_k_bar_stack[position_in_std_stack].TmIdx,
                     self.std_k_bar_stack[position_in_std_stack].H,
                     -1)
        elif method == 'L':
            p = Point(self.std_k_bar_stack[position_in_std_stack].TmIdx,
                     self.std_k_bar_stack[position_in_std_stack].L,
                     1)
        elif method == 'init':
            p = Point(self.std_k_bar_stack[position_in_std_stack].TmIdx,
                     self.std_k_bar_stack[position_in_std_stack].L,
                     0)
        return p
    
    def be_ripe(self):
        self.status == 1
    
        
```

```python
class SubTrend:
    def __init__(self,**kwargs):
        self.drt = kwargs['drt']
        self.TmIdx_start = kwargs['TmS']
        self.std_stack = [kwargs['']]
        self.status = 0
        
    def update(self, stick_std):
        if stick_std['drt'] != self.drt:
            flag2, new_stdK = self.std_stack[-1].update_T(*stick_std)
            if flag2 != 0:
                self.std_stack.append(StdK(*new_stdK))
        else:
            if len(self.std_stack) == 2:
                self.status = 1
        return self.status
            
```

```python
class Trend:
    def __init__(self, method, **kwargs):
        if method == 'init':
            self.TmIdx_start, self.TmIdx_end = kwargs['k_bar'][4], kwargs['k_bar'][4]
            self.O, self.C = kwargs['k_bar'][0], kwargs['k_bar'][3]
            self.stick_stack = [Stick('init', k_bar=kwargs['k_bar'])]
            self.peak = self.getPoint(0, 'start')
            self.peak_position_in_ss = 0
            self.sub_peak = []
            self.sub_peak_position_in_ss = 0
            self.drt = 0            
            self.upstd_stack, self.downstd_stack = [],[]
            self.status = 0 # 0:open; 1:ripe; 2:closed
            self.sub_trend_stack = []
        elif method == 'trim':
            pass
     
    def update(self, k_bar):
        self.TmIdx_end, self.C = k_bar[4], k_bar[3]
        # update_peak
        flag = 0 # 0:stain; 1: updated; 2:cloesd; 3create sub_trend 
        flag_tmp = self.updatedStick(k_bar)  # self.stick_stack & self.std_stack x 2 will be updated.
        if flag_tmp == 1: # update self.peak
            flag = self.updatePeak()          
            
        elif flag_tmp == 2:
            #  status ==2
            # flag = 2
            # new_trend = {}
            
            # sub_trend
            for st in self.sub_trend_stack:
                falg_st = st.update(self.stickToStdK(self.stick_stack[-2]))
                if flag_st == 1:
                    # subtrend :
                    # to merge trend
                    # flag = 2
                
            # 
            if self.updatePeak() == 1:
                flag = 1
            else: # peak, peak
                peak, pp= self.findPeak()
                # just for init: self.status==0 and peak!=0
                if self.status == 1: 
                    if peak != 0 and self.drt != peak:
                        self.status = 2
                        # self.peak = self.getPoint(pp, 'start')
                        #!!!!
                        self.peak_position_in_ss = pp
                    elif peak != 0 and self.drt == peak:
                        print('What the fuck!! ')
                        # but nothing to do.
                    elif peak == 0:
                        # peakss[-2]
                elif self.status == 2:
        elif flag_tmp == 3:
            drt_sub_trend = self.stick_stack[-1].drt
            tmp_std = self.sitckToStdK(self.stick_sack[-2])
            point_start = self.getPoint(-2, 'start')
            
            self.creatSubTrend()
        
        
        
            
            
        #!!!!!!!            
        if self.stick_stack[-1].drt == self.drt and len(self.stick_stack) > 5:
            flag = 3
            #!!!! trim
        return flag
    
    def updateStick(self, k_bar):
        flag = 0 # 0: stick_flag = 0; 1:stick_updated; 2: stick_flag = 2 and new std; 3: stick_flag=2 & 
        flag, new_stick = self.stick_stack[-1].update(k_bar) 
                       
        if flag == 2:
            self.stick_stack.append(Stick('trim', *new_stick))
            tmp_std = self.stickToStdK(self.stick_stack[-2])
        
            if len(self.stick_stack) == 1 and self.stick_stack[-2].drt == -1:
                self.upstd_stack.append(StdK(*tmp_std))
            elif len(self.stick_stack) == 1 and self.stick_stack[-2].drt == 1:
                self.downstd_stack.append(StdK(*tmp_std))
            elif len(self.stick_stack) > 1 and self.stick_stack[-2].drt == -1:
                flag2, new_stdK = self.upstd_stack[-1].update_T(*tmp_std)
                if flag2 != 0:
                    self.upstd_stack.append(StdK(*new_stdK))
                else:
                    flag = 3
            elif len(self.stick_stack) > 1 and self.stick_stack[-2].drt == 1:
                flag2, new_stdK = self.downstd_stack[-1].update_T(*tmp_std)
                if flag2 != 0:
                    self.downstd_stack.append(StdK(*new_stdK))
                else:
                    flag = 3
        
            
        return flag
    
    def findPeak(self):
        flag = 0
        if len(self.upstd_stack) > 2:
            if self.upstd_stack[-2].H > self.upstd_stack[-1].H and self.upstd_stack[-2].H > self.upstd_stack[-3].H:
                flag = -1
                peak_position = self.findIdxInSS(self.upstd_stack[-2].TmIdx)
        elif len(self.downstd_stack) > 2:
            if self.downstd_stack[-2].L < self.downstd_stack[-1].L and self.downstd_stack[-2].L < self.downstd_stack[-3].L:
                flag = 1
                peak_position = self.findIdxInSS(self.downstd_stack[-2].TmIdx)    
        return flag, peak_position
            
    def stickToStdK(self, stick_r):
        if stick_r.drt == 1:
            stdk = {'H':stick_r.peak_value, 'L':stick_r.O, 'TmIdx':stick_r.TmIdx_start, 'drt': 1}
        elif stick_r.drt == -1:
            stdk = {'L':stick_r.peak_value, 'H':stick_r.O, 'TmIdx':stick_r.TmIdx_start, 'drt': -1}
        return stdk
    
    def getPoint(self, position_in_stick_stack, method):
        if method == 'start':
            p = Point(self.stick_stack[position_in_stick_stack].TmIdx_start,
                     self.stick_stack[position_in_stick_stack].O,
                     self.stick_stack[position_in_stick_stack].drt)
        elif method == 'peak':
            p = Point(self.stick_stack[position_in_stick_stack].TmIdx_peak,
                     self.stick_stack[position_in_stick_stack].peak_value,
                     -self.stick_stack[position_in_stick_stack].drt)
        return p
    
    def updatePeak(self):
        flag = 0
        # status == 12
        if self.status == 0 or self.status == 1:
            if self.drt == self.stick_stack[-1].drt:
                if self.drt == 1 and self.peak.V <= self.stick_stack[-1].peak.V:
                    self.peak = getPoint(-1, 'peak')
                    flag = 1
                elif self.drt == -1 and self.peak.V >= self.stick_stack[-1].peak.V:
                    self.peak = getPoint(-1, 'peak')
                    flag = 1
        elif self.status == 2:
            if self.peak.drt ==1 self.stick_stack[-1].drt:
                if self.peak.drt == 1 and self.sub_peak.V <= self.stick_stack[-1].peak.V:
                    self.sub_peak = getPoint(-1, 'peak')
                    flag = 1
                elif self.peak.drt == -1 and self.sub_peak.V >= self.stick_stack[-1].peak.V:
                    self.sub_peak = getPoint(-1, 'peak')
                    flag = 1
        return flag
    
    def findIdxInSS(self, timeIndex):
        for i in range(len(self.stick_stack)):
            if self.stick_stack[i].TmIdx_start == timeIndex:
                return i
        
    
```
