发现mssql用的是Chinese_PRC_CI_AS，其中Chinese_PRC指针对大陆简体字的unicode, CI- 不区分大小写
AS - 区分重音


unicode 狭义指字符集，广义指字符集和编码方式；中文-2字节

utf-8,utf-16即编码方式，目的是网络传输，每个字符使用1-4个字节不等。 中文-3字节

字符串在Python内部的表示（python unicode object）是unicode编码(2字节)，因此，在做编码转换时，通常需要以unicode作为中间编码，即先将其他编码的字符串解码（decode）成unicode，再从unicode编码（encode）成另一种编码。 

其它“地区性编码”： ASCII(0-127) >> GC2312(6000汉字) >> GBK(windows cmd) >> GB18030(excel)
